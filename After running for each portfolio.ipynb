```python
def extract_values(row):
    pattern = r'(\w+)=(.*?)(?=\r|$)'
    key_value_pairs = re.findall(pattern, row['vars'])

    extracted_values = {}
    for key, value in key_value_pairs:
        if value.startswith("'") and value.endswith("'"):
            value = value[1:-1]
        extracted_values[key] = value

    return pd.Series(extracted_values)
```
