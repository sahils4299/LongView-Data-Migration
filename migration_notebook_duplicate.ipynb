```python
import re

def extract_address_components(short_address):
    # Define the regex pattern to capture the desired components
    pattern = r'^(\d+/)?(\d+)\s+([^,]+),\s+([^,]+),\s+([^,]+),\s+(\d+)$'

    # Match the pattern against the short_address
    matches = re.match(pattern, short_address)

    if matches:
        unit = matches.group(1) if matches.group(1) else ''  # Check if unit exists, assign empty string if not
        street_no = matches.group(2)
        street_1 = matches.group(3)
        street_2 = ''  # There is no street_2 in the given format, so assign empty string
        suburb = matches.group(4)
        state = matches.group(5)
        postcode = matches.group(6)

        return unit, street_no, street_1, street_2, suburb, state, postcode
    else:
        return '', '', '', '', '', '', ''  # Return empty strings if no match found

# Example usage
properties_before['Unit'], properties_before['Street No'], properties_before['Street 1'], properties_before['Street 2'], properties_before['Suburb'], properties_before['State'], properties_before['Postcode'] = zip(*properties_before['Short Address'].apply(extract_address_components))

```

```python
def extract_names(names_list):
    extracted_names = []

    for name in names_list:
        # Splitting the name string based on commas and "and"
        names = re.split(r',|\band\b', name)
        
        for name in names:
            # Removing salutations
            name_parts = [part.strip() for part in name.split() if part.lower() not in ['mr', 'mrs', 'miss']]
            full_name = ' '.join(name_parts)
            
            if full_name:
                extracted_names.append(full_name)

    return extracted_names
```
