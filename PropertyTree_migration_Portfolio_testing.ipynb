## 1. Preparing the properties

```python
"""
Function to call the properties using API from RPS 
Link : https://webservice.reapit.net.au/lng/rest/_doc/?ApiKey=c242c818604a72582f383ba23b2853d1&ClearCache=T#/Landlord/get_properties__ID__inspections

"""

import requests
from pandas.io.json import json_normalize

def get_property_details(prop_id):
    url = f"https://webservice.reapit.net.au/lng/rest/properties/rps_lng-{prop_id}"
    headers = {
        'ApiKey': "c242c818604a72582f383ba23b2853d1",
        "Accept": "application/json"
    }

    response = requests.get(url, headers=headers)
    if response.status_code == 200:
        return response.json()
    else:
        print(f"Error: {response.status_code} - {response.text}")
        return None
```

```python
# Creating a dataframe for Properties

df_properties = pd.DataFrame()
for prop_id in properties_before[properties_before['Agent Name'] == 'Kasey Neuendorf']['Reference'].tolist():
    result = get_property_details(prop_id)
    
    if result is not None:
        # Convert the result to a DataFrame
        df = json_normalize(result['Response'], sep='_')
        
        # Append the DataFrame to df_result
        df_properties = df_properties.append(df, ignore_index=True)
```

```python
df_properties[df_properties['Type'].isnull()]['HouseNumber']
```

### 1.1 Generating CSV for properties

```python
df_properties['Type'] = df_properties['Type'].apply(lambda x: x if str(x) =='nan' else x[0])
for index, row in df_properties.iterrows():
    if str(row['Type']) =='nan':
        df_properties.loc[index,'Type'] = 'Unit/Apartment' if '/' in row['HouseNumber'] else 'House'
```

```python
# Removing Property Type 'Car Space'
#df_properties = df_properties[df_properties['ID'] != 'rps_lng-CVA210304']
# if not df_properties['Type'].empty:
#     df_properties['Type'] = df_properties['Type'].explode()
df_properties['Type']  = df_properties['Type'].apply(lambda x: x.split('/')[1] if x =='Unit/Apartment' else x)
# Creating the lookup field for properties
df_properties['Type'] = df_properties['Type'].apply(lambda x: " Residential_" + x + " ")
property_rent_period_lookup = ["Daily", "Weekly", "Fortnightly", "Monthly", "Quarterly", "Yearly", "SquareMeterMonthly", "SquareMeterYearly"]
df_properties['RentalPeriod'] = df_properties['RentalPeriod'].apply(lambda x: next((match for match in property_rent_period_lookup if match.lower().__contains__(x)), x))
```
