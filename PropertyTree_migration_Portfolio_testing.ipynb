```python
# Create a dictionary to store the codes for each list
list_codes = {}

# Function to generate a 6-digit code for each list
def generate_list_code():
    return str((len(list_codes)+1) * 100000).zfill(6)

# Function to generate a code for each key in the list
def generate_key_code(list_code, index):
    if (index + 1) >= 10:
        return f"{list_code[0:4]}{str(index + 1)}"
    else:
        return f"{list_code[0:5]}{str(index + 1)}"
    
# Generate codes for each list
list_codes['financial_transactions'] = generate_list_code()
list_codes['property_maintenance'] = generate_list_code()
list_codes['fees_and_charges'] = generate_list_code()
list_codes['utilities_and_services'] = generate_list_code()
list_codes['legal_and_compliance'] = generate_list_code()
list_codes['miscellaneous'] = generate_list_code()
```

```python
# Generate codes for keys inside each list and update the code_description dictionary
for list_name, list_keys in [('financial_transactions', financial_transactions),
                             ('property_maintenance', property_maintenance),
                             ('fees_and_charges', fees_and_charges),
                             ('utilities_and_services', utilities_and_services),
                             ('legal_and_compliance', legal_and_compliance),
                             ('miscellaneous', miscellaneous)]:
    list_code = list_codes[list_name]
    for index, key in enumerate(list_keys):
        code_description[key] = generate_key_code(list_code, index)
```

## 1. Preparing the properties

```python
"""
Function to call the properties using API from RPS 
Link : https://webservice.reapit.net.au/lng/rest/_doc/?ApiKey=c242c818604a72582f383ba23b2853d1&ClearCache=T#/Landlord/get_properties__ID__inspections

"""

import requests
from pandas.io.json import json_normalize

def get_property_details(prop_id):
    url = f"https://webservice.reapit.net.au/lng/rest/properties/rps_lng-{prop_id}"
    headers = {
        'ApiKey': "c242c818604a72582f383ba23b2853d1",
        "Accept": "application/json"
    }

    response = requests.get(url, headers=headers)
    if response.status_code == 200:
        return response.json()
    else:
        print(f"Error: {response.status_code} - {response.text}")
        return None
```

```python
# Creating a dataframe for Properties

df_properties = pd.DataFrame()
for prop_id in properties_before[properties_before['Agent Name'] == 'Kasey Neuendorf']['Reference'].tolist():
    #print(prop_id)
    result = get_property_details(prop_id)
    
    if result is not None:
        # Convert the result to a DataFrame
        df = json_normalize(result['Response'], sep='_')
        
        # Append the DataFrame to df_result
        df_properties = df_properties.append(df, ignore_index=True)
```

### 1.1 Generating CSV for properties

```python
df_properties['Type'] = df_properties['Type'].apply(lambda x: x if str(x) =='nan' else x[0])
for index, row in df_properties.iterrows():
    if str(row['Type']) =='nan':
        df_properties.loc[index,'Type'] = 'Unit/Apartment' if '/' in row['HouseNumber'] else 'House'
```
