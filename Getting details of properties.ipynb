```python
#Importing the libraries
import pandas as pd
import requests
from pandas.io.json import json_normalize
```

```python
sold_props = pd.read_excel('Data Before/Portfolio Sale 1 - List.xlsx', sheet_name=None)
```

```python
prop_list = pd.read_excel('Data Before/Portfolio Sale 1 - List.xlsx', sheet_name='Sheet1')
prop_list_code = pd.read_excel('Data Before/Portfolio Sale 1 - List.xlsx', sheet_name='Sheet2')
```

```python
prop_list = prop_list.merge(prop_list_code,left_on= 'REAPIT Full Address',right_on='Short Address')[['REAPIT Full Address','Reference']]
```

```python
# https://webservice.reapit.net.au/lng/rest/properties/rps_lng-{prop_id}/tenancies
#https://webservice.reapit.net.au/lng/rest/properties/rps_lng-{prop_id}/worksOrders

def get_properties_transactions(prop_id):
    url = f"https://webservice.reapit.net.au/lng/rest/tenancies/{prop_id}/worksOrders"
    headers = {
        'ApiKey': "c242c818604a72582f383ba23b2853d1",
        "Accept": "application/json"
    }

    response = requests.get(url, headers=headers)
    if response.status_code == 200:
        return response.json()
    else:
        print(f"Error: {response.status_code} - {response.text}")
        return None
```

```python
# Getting the tenancies of the properties
df_prop_tenancies  = pd.DataFrame()

for prop_id in prop_list['Reference'].tolist():
    result = get_properties_transactions(prop_id)
    
    if result is not None:
        # Convert the result to a DataFrame
        df = json_normalize(result['Response'], sep='_')

        df.insert(0, 'Property Code', prop_id.split('_lng-')[1] if len(prop_id.split('_lng-')) > 1 else prop_id)
        
        # Append the DataFrame to df_result
        df_prop_tenancies = df_prop_tenancies.append(df, ignore_index=True)
```

```python
df_prop_tenancies = df_prop_tenancies.rename({'ID': 'Tenancy ID'},axis=1)[['Property Code','Tenancy ID']]
```

```python
df_properties_transactions = pd.DataFrame()
for index,row in df_prop_tenancies.iterrows():
    result = get_properties_transactions(row['Tenancy ID'])
    
    if result is not None:
        # Convert the result to a DataFrame
        df = json_normalize(result['Response'], sep='_')

        df.insert(0, 'Property Code', row['Property Code'])
        
        # Append the DataFrame to df_result
        df_properties_transactions = df_properties_transactions.append(df, ignore_index=True)
```
